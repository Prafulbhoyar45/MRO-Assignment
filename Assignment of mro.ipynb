### Q3. Create a table for Battery Maximum temperature in each Day also mention the Temperature sensor id (id=column title) for maximum temperature.


```python
# Find the maximum temperature and corresponding sensor ID for each day
max_temp_per_day = df.groupby(['Month', 'day'])[['Temp 1', 'Temp 2', 'Temp 3', 'Temp 4', 'Temp 5', 'Temp 6', 'Temp 7', 'Temp 8', 'Temp 9', 'Temp 10']].max()

# Find the column (sensor ID) with the maximum temperature for each day
max_temp_per_day['Max Sensor ID'] = max_temp_per_day.idxmax(axis=1)

# Rename the columns for clarity
max_temp_per_day.rename(columns={'Temp 1': 'Day 1', 'Temp 2': 'Day 2', 'Temp 3': 'Day 3', 'Temp 4': 'Day 4', 'Temp 5': 'Day 5',
                                 'Temp 6': 'Day 6', 'Temp 7': 'Day 7', 'Temp 8': 'Day 8', 'Temp 9': 'Day 9', 'Temp 10': 'Day 10'},
                        inplace=True)

# Reset the index for a cleaner table
max_temp_per_day.reset_index(inplace=True)

# Display the table
max_temp_per_day.head()
```

##### Temperature profile in discharge and charge state for last 7 days of month (Graphs) also represent a median temperature line in graph for each day. (x- axis = SOC; y-axis = Temperature) (SOC – Column AM)

```python
# Filter the data for the last 7 days of each month
last_7_days = df.groupby(['Month'])['day'].max()
filtered_df = df[df.apply(lambda x: x['day'] > (last_7_days[x['Month']] - 7), axis=1)]
```

```python
# Create separate DataFrames for discharge and charge states:
discharge_df = filtered_df[filtered_df['BMS State'] == 'Discharge']
charge_df = filtered_df[filtered_df['BMS State'] == 'Charge']
```

```python
# Calculate the median temperature for each day in both discharge and charge states:
median_temp_discharge = discharge_df.groupby(['Month', 'day', 'SOC'])['Temp 1'].median().reset_index()
median_temp_charge = charge_df.groupby(['Month', 'day', 'SOC'])['Temp 1'].median().reset_index()
```

```python
# Plot the temperature profiles with median lines using matplotlib and seaborn:
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots for discharge and charge states
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Discharge State Plot
sns.lineplot(x='SOC', y='Temp 1', data=discharge_df, hue='day', ax=axes[0])
sns.lineplot(x='SOC', y='Temp 1', data=median_temp_discharge, color='black', ax=axes[0])

axes[0].set_title('Temperature Profile in Discharge State')
axes[0].set_xlabel('SOC')
axes[0].set_ylabel('Temperature (°C)')

# Charge State Plot
sns.lineplot(x='SOC', y='Temp 1', data=charge_df, hue='day', ax=axes[1])
sns.lineplot(x='SOC', y='Temp 1', data=median_temp_charge, color='black', ax=axes[1])

axes[1].set_title('Temperature Profile in Charge State')
axes[1].set_xlabel('SOC')
axes[1].set_ylabel('Temperature (°C)')

# Set legend for the first plot (Discharge)
legend_labels = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7', 'Median']
axes[0].legend(legend_labels)

# Show the plots
plt.tight_layout()
plt.show()
```

### Q4. Create a Bar graph for maximum temperature achieved in each day of month (x axis – SOC, y axis – Temperature)

```python
# Find the maximum temperature for each day
max_temp_per_day = df.groupby(['Month', 'day'])[['Temp 1', 'Temp 2', 'Temp 3', 'Temp 4', 'Temp 5', 'Temp 6', 'Temp 7', 'Temp 8', 'Temp 9', 'Temp 10']].max()

# Calculate the average SOC for each day (assuming 'SOC' is a column in your dataset)
average_soc_per_day = df.groupby(['Month', 'day'])['SOC'].mean()

# Create a bar graph for the maximum temperature achieved in each day
plt.figure(figsize=(12, 6))
plt.bar(average_soc_per_day, max_temp_per_day.max(axis=1), width=0.8, alpha=0.7)
plt.xlabel('SOC')
plt.ylabel('Maximum Temperature (°C)')
plt.title('Maximum Temperature Achieved in Each Day of the Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Q5. All cell voltage graph for last 7 days in discharge and charge state

 -  separate graph for each day
 -  x- axis = SOC; y-axis = Voltage

```python
# Filter the data for the last 7 days
last_7_days = df.groupby(['Month'])['day'].max()
filtered_df = df[df.apply(lambda x: x['day'] > (last_7_days[x['Month']] - 7), axis=1)]

# Create separate graphs for each day
days = filtered_df['day'].unique()
for day in days:
    # Filter data for the specific day and Discharge state
    discharge_data = filtered_df[(filtered_df['day'] == day) & (filtered_df['BMS State'] == 'Discharge')]
    
    # Filter data for the specific day and Charge state
    charge_data = filtered_df[(filtered_df['day'] == day) & (filtered_df['BMS State'] == 'Charge')]
    
    # Create subplots for each cell voltage
    fig, axes = plt.subplots(4, 4, figsize=(15, 10))
    axes = axes.flatten()
    
    for i, cell in enumerate(range(1, 17)):
        # Discharge state cell voltage
        axes[i].plot(discharge_data['SOC'], discharge_data[f'Cell {cell}'], label='Discharge', color='blue')
        
        # Charge state cell voltage
        axes[i].plot(charge_data['SOC'], charge_data[f'Cell {cell}'], label='Charge', color='orange')
        
        axes[i].set_xlabel('SOC')
        axes[i].set_ylabel(f'Cell {cell} Voltage')
        axes[i].set_title(f'Cell {cell} Voltage on Day {day}')
        axes[i].legend()
    
    plt.tight_layout()
    plt.show()
```
